```python
import pandas as pd
import matplotlib.pyplot as plt

#%% data
url = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_US.csv'
covid_df = pd.read_csv(url, index_col=0)
```

```python
utah_data = covid_df[covid_df['Province_State'] == 'Utah']
date_columns = utah_data.columns[11:]  # Assuming the first 11 columns are metadata
county_names = utah_data['Admin2']
```

```python
# Highlight the county with most cases
highlight_county_idx = utah_data[date_columns].sum(axis=1).idxmax()
highlight_county_name = utah_data.loc[highlight_county_idx, 'Admin2']
highlight_county_data = utah_data.loc[highlight_county_idx, date_columns].diff().fillna(0)

# Plot all counties in grey
plt.figure(figsize=(15, 8))
for index, row in utah_data.iterrows():
    daily_cases = row[date_columns].diff().fillna(0)
    plt.plot(date_columns, daily_cases, color='grey', alpha=0.5, linewidth=0.7)

# Plot the highlighted county in orange
plt.plot(date_columns, highlight_county_data, color='orange', alpha = 0.5,\
         label=f"{highlight_county_name} (Highlighted)", linewidth=0.7)

# Format dates
plt.xticks(range(0, len(date_columns), 60), pd.to_datetime(date_columns).strftime('%Y-%m')[::60], rotation=45)

plt.title('COVID-19 Cases in Utah Counties', fontsize=14)
plt.xlabel('Date', fontsize=12)
plt.ylabel('New cases', fontsize=12)
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# florida
df_flo = covid_df[covid_df['Province_State'] == 'Florida']
flo_data = df_flo.iloc[:,11:]
flo_max_data = flo_data.loc[flo_data.sum(axis = 1).idxmax(),:]
flo_max_row = df_flo.loc[flo_data.sum(axis = 1).idxmax(),:]
```
