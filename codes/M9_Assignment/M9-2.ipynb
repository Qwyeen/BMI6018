```python
# Question 3
def interchange_columns(df, col1, col2):
    columns = list(df.columns)
    idx1, idx2 = columns.index(col1), columns.index(col2) 
    columns[idx1], columns[idx2] = columns[idx2], columns[idx1]
    output = df[columns]
    return(output)


df = pd.DataFrame(np.arange(20).reshape(-1, 5), columns=list('abcde'))
df_ac_swap = interchange_columns(df, 'a', 'c')
```

```python
# Question 5
df = pd.DataFrame(np.random.randint(1,100, 40).reshape(10, -1),
                  columns=list('pqrs'), index=list('abcdefghij'))
```

```python
nearest_row = []
dist = []
for index,row in df.iterrows():
    ls_distance = ((row-df)**2).sum(axis = 1)**0.5
    ls_distance[index] = ls_distance.max() # make sure not to choose its own row
    nearest_row.append(ls_distance.idxmin())
    dist.append(ls_distance.min())
df['nearest_row'] = nearest_row
df['dist'] = dist
```
